# Filtrado de filas

Cual es el registro del pasajero más viejo que viajaba en el Titanic ?

```python
df[df["Age"] == df["Age"].max()]
```

Muestra los pasajeros que sobrevivieron y eran mayores de 60 años

```python
df[(df["Survived"] == 1) & (df["Age"] > 60)]
```

Muestra los pasajeros que tuvieran algun familiar y no sobrevivieron 

```python
df[((df["Parents/Children Aboard"] > 0) | (df["Siblings/Spouses Aboard"] > 0)) & (df["Survived"] == 0)]
```

Muestra un dataframe solo con las columnas que acaben en e

```python
df.filter(regex = "e$", axis=1) # axis = 1 para columnas, axis = 0 para filas
```

Muestra un dataframe solo con las columnas que tengas "Mary" en el nombre

```python
df.filter(like = "Mary", axis = 0) # El filtro se hace por los índices. Como son numéricos, no se encuentra nada
```

# Limpiar Datos

```python
import numpy as np
```

```python
df_ = pd.DataFrame({'x':[1,2,np.nan,np.nan, 5]})
df_
```

Podemos rellenar los datos absentes con un 0

```python
df_.fillna(value = 0)
```

Rellenar los datos NaN con el valor anterior o posterior

```python
df_.fillna(method="bfill") # Lo rellena con el valor de la posición siguiente
```

```python
df_.fillna(method="ffill") # Lo rellena con el valor de la posición anterior
```

Podemos hacer una interpolación 


```python
df_.interpolate()
```

Eliminar los datos que faltan

```python
df_.dropna()
```
